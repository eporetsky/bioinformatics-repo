# Converting the Phytozome GFF3 to a table of gene coordinates

```python
df = pd.read_csv("Zmays_493_RefGen_V4.gene.gff3", sep="\t", header=None, skiprows=3, low_memory=False)
df.head()
```

```python
genes = df[df[2]=="gene"]

# Trying to avoid error message about copy of a slice from a DataFrame saved onto itself
#https://stackoverflow.com/questions/51344329/pandas-try-using-locrow-indexer-col-indexer-value-instead
pd.options.mode.chained_assignment = None

def gff_split(val):
    return val.split("=")[-1]

genes.iloc[:,8] = genes[8].apply(gff_split)
genes = genes.drop(columns=[1, 2, 5, 7], axis=1)
genes.columns = ["chr", "start", "end", "strand", "id"]
genes
```
